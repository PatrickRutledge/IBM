# TASK 3: Calculate the distances between a launch site to its proximities


Next, we need to explore and analyze the proximities of launch sites.


Let's first add a `MousePosition` on the map to get coordinate for a mouse over a point on the map. As such, while you are exploring the map, you can easily find the coordinates of any points of interests (such as railway)


```python
"""# Add Mouse Position to get the coordinate (Lat, Long) for a mouse over on the map
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)

site_map.add_child(mouse_position)
site_map
"""
import folium
from folium.plugins import MousePosition, MarkerCluster # Ensure MarkerCluster is imported if you use it
import pandas as pd
import numpy as np

# This code assumes spacex_df (with 'Lat', 'Long', 'class', 'Launch Site', and 'marker_color' columns)
# is already defined in your Python environment from a previous step (e.g., loading from CSV and processing).
# Please ensure your spacex_df DataFrame has your complete data before running this.

# Function to assign color based on launch outcome
def assign_marker_color(launch_outcome):
    return 'green' if launch_outcome == 1 else 'red'

# Create the 'marker_color' column in your spacex_df.
# This line should be run after your spacex_df is loaded and before the map generation.
spacex_df['marker_color'] = spacex_df['class'].apply(assign_marker_color)


# --- 2. Initialize the Folium Map ---
# Calculate the mean latitude and longitude from your spacex_df to center the map.
# This helps ensure a good initial view that tries to show all your launch sites.
center_lat = spacex_df['Lat'].mean()
center_long = spacex_df['Long'].mean()

# Create the main folium map object.
# A zoom_start of 4 or 5 is generally good for viewing the continental US,
# which should encompass all your major launch site clusters.
site_map = folium.Map(location=[center_lat, center_long], zoom_start=4)


# --- 3. Add your existing map points (markers/clusters) ---
# Initialize the MarkerCluster plugin.
# disableClusteringAtZoom=18 will ensure individual markers appear when you zoom in very close.
marker_cluster = MarkerCluster(disableClusteringAtZoom=18)

# Add the MarkerCluster layer to your map.
site_map.add_child(marker_cluster)

# Loop through each row in your spacex_df DataFrame.
# For each launch record, a Folium marker is created and added to the cluster.
for index, record in spacex_df.iterrows():
    # Get coordinates and color for the current launch record.
    marker_coords = [record['Lat'], record['Long']]
    marker_color = record['marker_color']
    launch_site_name = record['Launch Site']
    launch_outcome_class = record['class']

    # Create a Marker object.
    # The icon color will be green for success (class=1) and red for failure (class=0).
    # The 'info-sign' icon provides a distinct visual element.
    # The popup provides useful information when the marker is clicked.
    marker = folium.Marker(
        location=marker_coords,
        icon=folium.Icon(color='white', icon_color=marker_color, icon='info-sign'),
        popup=f"Launch Site: {launch_site_name}<br>"
              f"Outcome: {'Success' if launch_outcome_class == 1 else 'Failure'}"
    )
    # Add the created marker to the marker_cluster.
    marker_cluster.add_child(marker)

# If you had other individual markers or circles that are not part of the clustering,
# you would add them here to site_map directly.
# Example: folium.Circle(location=[lat, lon], radius=...).add_to(site_map)
# Example: folium.Marker(location=[lat, lon], popup="Something else").add_to(site_map)


# --- 4. Add the Mouse Position plugin ---
# This JavaScript function formats the numbers for display.
formatter = "function(num) {return L.Util.formatNum(num, 5);};"

# Create the MousePosition plugin object.
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=5, # Number of digits after decimal point for display
    prefix='Lat:',
    lat_formatter=formatter, # Apply the custom formatter
    lng_formatter=formatter, # Apply the custom formatter
)
# Add the MousePosition plugin to your map.
site_map.add_child(mouse_position)


# --- 5. Display the Map (once, at the end of the combined block) ---
# This command renders the interactive Folium map in your output cell.
site_map
```

Now zoom in to a launch site and explore its proximity to see if you can easily find any railway, highway, coastline, etc. Move your mouse to these points and mark down their coordinates (shown on the top-left) in order to the distance to the launch site.


You can calculate the distance between two points on the map based on their `Lat` and `Long` values using the following method:


```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
